```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

```

```python
url="https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY"
response=(requests.get(url))
response
```

```python
soup=BeautifulSoup(response.content,"html.parser")
```

```python
Images=[]
image=soup.find_all("img",class_="_396cs4 _3exPp9")
for i in image:
    Images.append(i.get("src"))
        
```

```python
Name=[]
name=soup.find_all("div",class_="_4rR01T")
for i in name:
    Name.append(i.text)
Name
```

```python
Rating=[]
rating=soup.find_all("div",class_="_3LWZlK")
for i in rating:
    Rating.append(i.text)
Rating
```

```python
Price=[]
price=soup.find_all("div",class_="_25b18c")
for i in price:
    Price.append(i.text)
```

```python
df=pd.DataFrame()
```

```python
df=pd.DataFrame()
df["NAME"]=Name
df["PRICE"]=Price
df["RATING"]=Rating
df["IMAGE"]=Images
df
```
